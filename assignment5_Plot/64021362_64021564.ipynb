1. Plot relation for each feature in subplot (maximum 8 subplots)

```python
# !pip install matplotlib seaborn pandas

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

current_directory = os.getcwd()
print(current_directory)

file_name = 'Weather Data.csv'
file_path = os.path.join(current_directory, file_name)

iris = pd.read_csv(file_path)


print("Weather Data:")
print("For viewing weather conditions.")
print(iris.info())
print("From https://www.kaggle.com/datasets/vikramkathare/weather-dataset/")

total_samples = len(iris)
print("\nTotal Number of Samples:", total_samples)

total_classes_column = 'Weather'
total_classes = len(iris[total_classes_column].unique())
print("Total Number of Classes:", total_classes)


fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(16, 8))

for i, column in enumerate(iris.columns[:-1]):
    sns.histplot(iris, x=column, kde=True, ax=axes[i//4, i%4])

plt.show()
```
